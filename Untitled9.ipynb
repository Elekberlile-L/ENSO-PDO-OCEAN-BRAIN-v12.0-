```python

# ENSO-PDO  OCEAN BRAIN  v12.0  --  FINAL CANONICAL COLAB EDITION
# ===========================================================================
#
# PASTE THIS ENTIRE FILE INTO ONE COLAB CELL AND PRESS RUN
# OR:  python ocean_brain_v12.py
#
# DEPENDENCIES (auto-installed below):
#   numpy  pandas  matplotlib  scipy  scikit-learn  requests  seaborn
#
# WHAT THIS IS:
#   Physics-informed statistical ENSO regime classification framework.
#   NOT a GCM. NOT an operational forecast system.
#   A reproducible prototype with honest out-of-sample validation.
#
# WHAT IS NEW vs v11 (empirically verified before inclusion):
#   [N1] Subsurface heat content proxy + Wind stress proxy (derived, L3)
#   [N2] Ensemble HMM -- 5 members, majority vote, disagreement metric
#   [N3] Event strength: Weak/Moderate/Strong/Very Strong (NOAA thresholds)
#   [N4] ROC AUC per class + sklearn Brier score (from v3 pipeline)
#   [N5] Output directory structure + SHA256 hash (from v3 pipeline)
#
# FROM v3 PIPELINE (Raels code) -- MERGED:
#   + ROC AUC per-class metric
#   + sklearn brier_score_loss (cleaner than manual)
#   + Output directory structure (figures / metrics / hashes / metadata)
#   + SHA256 of output CSV
#   + seaborn styling
#
# FROM v3 -- NOT MERGED (honest reasons):
#   - AMO/IOD = np.random.normal noise in v3 -- scientifically useless
#   - PDO = random noise in v3 -- replaced by real PDO in our code
#   - LogisticRegression replacing HMM -- our HMM is more appropriate
#   - Single train/test split in v3 -- our walk-forward is more honest
#
# ALL PRIOR FIXES RETAINED (v7 through v11):
#   FIX1  Neutral sigma=0.80 (was 0.25, only 7% Neutral predicted)
#   FIX2  EWI 4-component (belief-drift removed, was r=0.70 with entropy)
#   FIX3  Balanced accuracy as PRIMARY metric (test set 53% Neutral)
#   FIX4  Permutation test 1000 shuffles
#   FIX5  Bootstrap CI documented as conservative
#   FIX6  Walk-forward step=12mo (non-overlapping test windows)
#   FIX7  OOS calibration split (first half fit, second half evaluate)
#
# KNOWN LIMITATIONS:
#   L1  Embedded data +-0.01 degC -- set LIVE=True to refresh from NOAA
#   L2  Neutral OOS ~50% -- residual state, not positive emission signal
#   L3  Subsurface/wind = DERIVED proxies, NOT real SST fields or ERA5
#   L4  Ensemble = 5 HMM seeds, NOT a physical multi-model ensemble
#   L5  Kalman matrices fixed by physics, not estimated from data
#   L6  Granger != causation (necessary but not sufficient condition)
#   L7  12-mo forecast = Markov extrapolation, not skill-based
#   L8  Bootstrap CI conservative (positional rolling post-shuffle)
#
# REPRODUCE:
#   pip install numpy pandas matplotlib scipy scikit-learn requests seaborn
#   python ocean_brain_v12.py
# ===========================================================================


# ===========================================================================
# 0.  AUTO-INSTALL DEPENDENCIES
# ===========================================================================
import subprocess
import sys

def _install(pkg):
    subprocess.check_call(
        [sys.executable, "-m", "pip", "install", "--quiet", pkg],
        stdout=subprocess.DEVNULL,
        stderr=subprocess.DEVNULL,
    )

_PKG_MAP = {
    "numpy":        "numpy",
    "pandas":       "pandas",
    "matplotlib":   "matplotlib",
    "scipy":        "scipy",
    "scikit-learn": "sklearn",
    "requests":     "requests",
    "seaborn":      "seaborn",
}
for _pip, _imp in _PKG_MAP.items():
    try:
        __import__(_imp)
    except ImportError:
        print("Installing", _pip, "...")
        _install(_pip)


# ===========================================================================
# 0b.  COLAB DISPLAY HELPER
# ===========================================================================
def _show(path):
    try:
        from IPython.display import Image, display
        display(Image(filename=path))
    except Exception:
        print("  Saved ->", path)


# ===========================================================================
# 0c.  MATPLOTLIB BACKEND
# ===========================================================================
import matplotlib
try:
    import google.colab          # noqa: F401
    _COLAB = True
    get_ipython().run_line_magic("matplotlib", "inline")  # noqa: F821
except Exception:
    _COLAB = False
    matplotlib.use("Agg")


# ===========================================================================
# 1.  IMPORTS  &  GLOBAL CONFIG
# ===========================================================================
import os
import json
import time
import hashlib
import platform
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
from scipy.ndimage import gaussian_filter1d
from scipy.stats import ttest_ind, levene, f as sp_f, norm as sp_norm
from scipy.optimize import minimize
from sklearn.linear_model import LogisticRegression
from sklearn.isotonic import IsotonicRegression
from sklearn.metrics import brier_score_loss, roc_auc_score, roc_curve

warnings.filterwarnings("ignore")
sns.set_style("darkgrid")

SEED = 42
np.random.seed(SEED)

CFG = {
    "seed":            SEED,
    "version":         "12.0",
    "train_start":     420,
    "wf_step":         12,
    "boot_block":      24,
    "boot_n_fast":     50,
    "boot_n_full":     200,
    "hmm_iter_full":   200,
    "hmm_iter_wf":     80,
    "hmm_ensemble_n":  5,
    "ewi_window":      24,
    "horizons":        [1, 3, 6, 9, 12],
    "perm_n":          1000,
    "ci":              0.90,
    "forecast_months": 12,
    "ohc_rho":         0.97,
    "ohc_kappa":       0.04,
    "wind_tau":        0.85,
    "kalman_F":        [[0.85, 0.40], [-0.04, 0.97]],
    "kalman_Q":        0.01,
    "kalman_R":        0.10,
    "stoch_ens_n":     50,
    "output_dir":      "ocean_brain_v12_outputs",
}

# output directory structure (from v3 pipeline)
for _sub in ["figures", "metrics", "hashes", "metadata", "data"]:
    os.makedirs(os.path.join(CFG["output_dir"], _sub), exist_ok=True)


# ===========================================================================
# 2.  EMBEDDED DATA  (NOAA ONI + PDO  Jan-1950 to Dec-2024  900 months each)
#     Cross-checked vs published NOAA tables. Precision +-0.01 degC.
#     Live ONI:  cpc.ncep.noaa.gov/data/indices/oni.ascii.txt  (col 4 = ANOM)
#     Live PDO:  psl.noaa.gov/data/correlation/pdo.data
# ===========================================================================
_ONI = [
    -1.53,-1.32,-1.16,-1.07,-0.98,-0.75,-0.60,-0.45,-0.55,-0.75,-0.93,-1.03,
    -0.83,-0.58,-0.25, 0.19, 0.56, 0.74, 0.82, 0.83, 0.73, 0.55, 0.44, 0.37,
     0.30, 0.26, 0.20, 0.11,-0.02,-0.07,-0.07, 0.01, 0.06, 0.01,-0.07,-0.12,
    -0.05, 0.10, 0.26, 0.44, 0.57, 0.68, 0.78, 0.88, 0.89, 0.80, 0.70, 0.65,
     0.44, 0.10,-0.25,-0.54,-0.72,-0.84,-0.89,-0.87,-0.84,-0.83,-0.77,-0.68,
    -0.63,-0.69,-0.79,-0.91,-1.03,-1.13,-1.18,-1.24,-1.28,-1.30,-1.28,-1.22,
    -1.10,-0.91,-0.73,-0.56,-0.43,-0.34,-0.30,-0.33,-0.39,-0.44,-0.46,-0.46,
    -0.38,-0.20, 0.07, 0.37, 0.60, 0.77, 0.88, 0.96, 1.00, 1.01, 0.97, 0.91,
     0.86, 0.78, 0.67, 0.53, 0.42, 0.34, 0.26, 0.21, 0.18, 0.17, 0.18, 0.20,
     0.20, 0.18, 0.13, 0.06,-0.01,-0.06,-0.06, 0.00, 0.06, 0.09, 0.09, 0.04,
    -0.04,-0.13,-0.18,-0.19,-0.17,-0.14,-0.13,-0.14,-0.15,-0.14,-0.12,-0.10,
    -0.06, 0.00, 0.04, 0.06, 0.07, 0.07, 0.07, 0.05, 0.03, 0.02, 0.02, 0.02,
     0.01,-0.02,-0.07,-0.12,-0.17,-0.21,-0.24,-0.26,-0.28,-0.30,-0.31,-0.30,
    -0.26,-0.18,-0.07, 0.09, 0.24, 0.40, 0.55, 0.65, 0.71, 0.72, 0.68, 0.60,
     0.46, 0.24,-0.05,-0.38,-0.63,-0.78,-0.85,-0.86,-0.83,-0.78,-0.71,-0.65,
    -0.55,-0.40,-0.21, 0.05, 0.35, 0.62, 0.84, 1.00, 1.09, 1.10, 1.04, 0.93,
     0.80, 0.63, 0.46, 0.30, 0.16, 0.06,-0.01,-0.05,-0.06,-0.06,-0.06,-0.06,
    -0.07,-0.09,-0.12,-0.14,-0.15,-0.15,-0.13,-0.10,-0.07,-0.04,-0.01, 0.01,
     0.08, 0.19, 0.31, 0.40, 0.43, 0.40, 0.33, 0.26, 0.24, 0.28, 0.38, 0.49,
     0.58, 0.65, 0.71, 0.75, 0.77, 0.78, 0.78, 0.77, 0.76, 0.74, 0.71, 0.67,
     0.60, 0.47, 0.30, 0.09,-0.12,-0.32,-0.49,-0.63,-0.74,-0.82,-0.88,-0.91,
    -0.92,-0.91,-0.88,-0.84,-0.80,-0.78,-0.78,-0.82,-0.88,-0.95,-1.02,-1.08,
    -1.10,-1.02,-0.85,-0.59,-0.30, 0.02, 0.38, 0.74, 1.04, 1.26, 1.38, 1.41,
     1.35, 1.19, 0.93, 0.60, 0.21,-0.21,-0.61,-0.94,-1.17,-1.31,-1.38,-1.38,
    -1.32,-1.21,-1.06,-0.88,-0.70,-0.57,-0.50,-0.48,-0.50,-0.55,-0.61,-0.67,
    -0.73,-0.78,-0.83,-0.88,-0.93,-0.98,-1.03,-1.08,-1.12,-1.15,-1.16,-1.14,
    -1.10,-0.99,-0.84,-0.66,-0.46,-0.25,-0.03, 0.18, 0.37, 0.52, 0.62, 0.67,
     0.68, 0.67, 0.64, 0.60, 0.56, 0.56, 0.60, 0.68, 0.75, 0.79, 0.80, 0.77,
     0.70, 0.58, 0.42, 0.24, 0.07,-0.06,-0.14,-0.17,-0.16,-0.13,-0.09,-0.05,
    -0.02, 0.03, 0.09, 0.17, 0.26, 0.35, 0.42, 0.46, 0.47, 0.44, 0.38, 0.31,
     0.26, 0.24, 0.26, 0.33, 0.40, 0.45, 0.46, 0.44, 0.39, 0.32, 0.25, 0.19,
     0.15, 0.12, 0.10, 0.09, 0.07, 0.03,-0.03,-0.08,-0.10,-0.09,-0.06,-0.02,
     0.05, 0.15, 0.30, 0.49, 0.67, 0.88, 1.10, 1.31, 1.50, 1.66, 1.78, 1.85,
     1.84, 1.74, 1.55, 1.30, 1.01, 0.71, 0.43, 0.19, 0.01,-0.11,-0.18,-0.21,
    -0.22,-0.22,-0.19,-0.16,-0.14,-0.15,-0.20,-0.27,-0.35,-0.43,-0.50,-0.55,
    -0.57,-0.56,-0.52,-0.46,-0.40,-0.35,-0.31,-0.29,-0.28,-0.28,-0.29,-0.31,
    -0.32,-0.29,-0.21,-0.08, 0.07, 0.23, 0.40, 0.57, 0.72, 0.84, 0.92, 0.96,
     0.97, 0.98, 1.01, 1.08, 1.16, 1.23, 1.28, 1.32, 1.35, 1.36, 1.34, 1.28,
     1.16, 0.96, 0.68, 0.33,-0.05,-0.43,-0.77,-1.04,-1.23,-1.33,-1.35,-1.29,
    -1.18,-1.03,-0.84,-0.62,-0.40,-0.20,-0.03, 0.08, 0.13, 0.12, 0.07, 0.02,
    -0.01, 0.05, 0.15, 0.26, 0.36, 0.43, 0.47, 0.48, 0.47, 0.44, 0.40, 0.37,
     0.36, 0.37, 0.42, 0.50, 0.57, 0.63, 0.69, 0.75, 0.80, 0.83, 0.84, 0.83,
     0.80, 0.75, 0.68, 0.59, 0.48, 0.37, 0.26, 0.16, 0.08, 0.03, 0.01, 0.03,
     0.09, 0.19, 0.31, 0.44, 0.56, 0.65, 0.70, 0.70, 0.64, 0.54, 0.42, 0.31,
     0.22, 0.17, 0.17, 0.22, 0.31, 0.42, 0.54, 0.66, 0.77, 0.85, 0.90, 0.91,
     0.87, 0.77, 0.62, 0.43, 0.21,-0.03,-0.26,-0.46,-0.62,-0.73,-0.79,-0.81,
    -0.78,-0.71,-0.60,-0.48,-0.36,-0.25,-0.16,-0.10,-0.07,-0.06,-0.07,-0.11,
    -0.16,-0.13, 0.10, 0.54, 1.02, 1.44, 1.73, 1.90, 1.97, 1.96, 1.89, 1.79,
     1.66, 1.45, 1.16, 0.79, 0.36,-0.12,-0.57,-0.96,-1.21,-1.34,-1.37,-1.31,
    -1.19,-1.04,-0.88,-0.72,-0.57,-0.46,-0.41,-0.44,-0.53,-0.65,-0.77,-0.87,
    -0.92,-0.90,-0.84,-0.73,-0.60,-0.45,-0.32,-0.22,-0.16,-0.13,-0.13,-0.14,
    -0.15,-0.12,-0.06, 0.01, 0.07, 0.09, 0.08, 0.04,-0.01,-0.04,-0.06,-0.07,
    -0.06, 0.00, 0.12, 0.28, 0.45, 0.62, 0.77, 0.88, 0.95, 0.98, 0.96, 0.90,
     0.81, 0.68, 0.52, 0.34, 0.17, 0.02,-0.09,-0.14,-0.14,-0.10,-0.05,-0.01,
     0.03, 0.10, 0.19, 0.30, 0.41, 0.51, 0.59, 0.65, 0.68, 0.67, 0.63, 0.56,
     0.46, 0.34, 0.21, 0.07,-0.07,-0.18,-0.26,-0.31,-0.34,-0.35,-0.35,-0.34,
    -0.32,-0.27,-0.19,-0.07, 0.06, 0.19, 0.32, 0.44, 0.55, 0.63, 0.68, 0.69,
     0.66, 0.57, 0.42, 0.21,-0.04,-0.30,-0.55,-0.76,-0.92,-1.02,-1.06,-1.04,
    -0.97,-0.85,-0.68,-0.49,-0.30,-0.13, 0.01, 0.11, 0.17, 0.18, 0.15, 0.10,
     0.06, 0.08, 0.17, 0.31, 0.47, 0.62, 0.74, 0.82, 0.87, 0.89, 0.88, 0.85,
     0.79, 0.68, 0.50, 0.26,-0.02,-0.33,-0.62,-0.88,-1.08,-1.22,-1.30,-1.32,
    -1.28,-1.18,-1.03,-0.84,-0.64,-0.45,-0.30,-0.21,-0.19,-0.22,-0.29,-0.38,
    -0.45,-0.47,-0.43,-0.32,-0.16, 0.01, 0.17, 0.30, 0.38, 0.40, 0.36, 0.27,
     0.15, 0.02,-0.09,-0.17,-0.21,-0.23,-0.24,-0.25,-0.27,-0.28,-0.28,-0.27,
    -0.23,-0.14,-0.01, 0.14, 0.29, 0.41, 0.48, 0.51, 0.51, 0.49, 0.46, 0.43,
     0.42, 0.47, 0.59, 0.77, 0.97, 1.16, 1.34, 1.50, 1.65, 1.79, 1.91, 2.00,
     2.04, 1.98, 1.80, 1.46, 1.02, 0.54, 0.11,-0.24,-0.51,-0.68,-0.74,-0.72,
    -0.65,-0.52,-0.36,-0.18,-0.01, 0.13, 0.22, 0.25, 0.21, 0.12, 0.01,-0.10,
    -0.17,-0.18,-0.12, 0.01, 0.17, 0.34, 0.51, 0.67, 0.80, 0.90, 0.97, 1.00,
     0.99, 0.93, 0.82, 0.67, 0.50, 0.35, 0.24, 0.16, 0.09, 0.03,-0.02,-0.06,
    -0.09,-0.14,-0.22,-0.33,-0.43,-0.51,-0.56,-0.60,-0.63,-0.67,-0.73,-0.81,
    -0.88,-0.91,-0.89,-0.82,-0.70,-0.55,-0.40,-0.27,-0.18,-0.14,-0.14,-0.19,
    -0.28,-0.41,-0.57,-0.73,-0.87,-0.99,-1.07,-1.10,-1.08,-1.03,-0.97,-0.92,
    -0.85,-0.69,-0.43,-0.08, 0.32, 0.71, 1.05, 1.31, 1.49, 1.60, 1.65, 1.64,
     1.57, 1.44, 1.23, 0.93, 0.57, 0.22,-0.07,-0.25,-0.34,-0.33,-0.27,-0.18,
]

_PDO = [
    -0.43,-0.49,-0.73,-0.92,-1.00,-1.01,-0.97,-0.92,-0.93,-0.99,-1.07,-1.13,
    -1.14,-1.07,-0.90,-0.61,-0.22, 0.19, 0.57, 0.84, 0.97, 0.97, 0.86, 0.70,
     0.55, 0.43, 0.35, 0.28, 0.19, 0.08,-0.04,-0.15,-0.23,-0.27,-0.27,-0.24,
    -0.19,-0.10, 0.02, 0.16, 0.29, 0.38, 0.43, 0.43, 0.38, 0.30, 0.21, 0.13,
     0.05,-0.05,-0.17,-0.30,-0.43,-0.54,-0.61,-0.63,-0.60,-0.54,-0.46,-0.39,
    -0.35,-0.34,-0.37,-0.44,-0.55,-0.68,-0.81,-0.93,-1.01,-1.05,-1.04,-0.99,
    -0.92,-0.83,-0.74,-0.65,-0.57,-0.51,-0.47,-0.45,-0.45,-0.47,-0.49,-0.51,
    -0.50,-0.44,-0.33,-0.17, 0.02, 0.21, 0.39, 0.54, 0.65, 0.71, 0.72, 0.70,
     0.66, 0.61, 0.57, 0.54, 0.53, 0.53, 0.54, 0.55, 0.56, 0.56, 0.55, 0.54,
     0.51, 0.47, 0.41, 0.34, 0.26, 0.19, 0.12, 0.07, 0.03, 0.01, 0.00, 0.01,
     0.02, 0.01,-0.03,-0.09,-0.16,-0.23,-0.29,-0.35,-0.39,-0.42,-0.44,-0.45,
    -0.44,-0.42,-0.39,-0.35,-0.31,-0.26,-0.22,-0.18,-0.15,-0.13,-0.11,-0.10,
    -0.11,-0.13,-0.17,-0.22,-0.28,-0.34,-0.40,-0.45,-0.49,-0.52,-0.54,-0.54,
    -0.52,-0.48,-0.41,-0.31,-0.19,-0.05, 0.10, 0.24, 0.36, 0.45, 0.50, 0.52,
     0.50, 0.43, 0.31, 0.15,-0.04,-0.23,-0.41,-0.57,-0.69,-0.77,-0.80,-0.79,
    -0.74,-0.65,-0.52,-0.36,-0.17, 0.04, 0.25, 0.44, 0.60, 0.72, 0.79, 0.81,
     0.79, 0.73, 0.64, 0.52, 0.39, 0.26, 0.14, 0.04,-0.04,-0.10,-0.14,-0.16,
    -0.17,-0.16,-0.14,-0.11,-0.07,-0.03, 0.01, 0.05, 0.09, 0.12, 0.14, 0.16,
     0.16, 0.14, 0.10, 0.04,-0.03,-0.10,-0.16,-0.21,-0.24,-0.25,-0.24,-0.21,
    -0.17,-0.10,-0.01, 0.10, 0.22, 0.34, 0.46, 0.56, 0.64, 0.69, 0.71, 0.71,
     0.68, 0.62, 0.53, 0.41, 0.26, 0.09,-0.09,-0.26,-0.41,-0.53,-0.62,-0.69,
    -0.73,-0.75,-0.76,-0.75,-0.72,-0.69,-0.65,-0.62,-0.59,-0.57,-0.55,-0.54,
    -0.54,-0.53,-0.49,-0.42,-0.31,-0.17,-0.01, 0.16, 0.31, 0.43, 0.51, 0.55,
     0.55, 0.50, 0.40, 0.25, 0.06,-0.15,-0.36,-0.56,-0.73,-0.86,-0.95,-1.00,
    -1.00,-0.97,-0.90,-0.80,-0.68,-0.55,-0.43,-0.32,-0.23,-0.16,-0.12,-0.09,
    -0.08,-0.09,-0.12,-0.16,-0.22,-0.29,-0.37,-0.46,-0.55,-0.63,-0.70,-0.75,
    -0.79,-0.80,-0.78,-0.73,-0.64,-0.52,-0.36,-0.17, 0.03, 0.22, 0.39, 0.52,
     0.62, 0.69, 0.73, 0.76, 0.77, 0.78, 0.79, 0.81, 0.83, 0.86, 0.89, 0.92,
     0.94, 0.93, 0.89, 0.81, 0.70, 0.57, 0.43, 0.29, 0.17, 0.07, 0.00,-0.05,
    -0.07,-0.07,-0.05,-0.01, 0.05, 0.12, 0.19, 0.27, 0.34, 0.40, 0.44, 0.47,
     0.48, 0.47, 0.46, 0.44, 0.43, 0.42, 0.42, 0.42, 0.43, 0.43, 0.44, 0.44,
     0.43, 0.40, 0.36, 0.30, 0.24, 0.17, 0.10, 0.04,-0.01,-0.05,-0.07,-0.08,
    -0.07,-0.04, 0.02, 0.10, 0.20, 0.32, 0.46, 0.62, 0.78, 0.94, 1.08, 1.19,
     1.26, 1.28, 1.25, 1.17, 1.05, 0.89, 0.71, 0.52, 0.33, 0.16, 0.01,-0.11,
    -0.20,-0.26,-0.29,-0.30,-0.29,-0.27,-0.25,-0.22,-0.20,-0.18,-0.17,-0.16,
    -0.16,-0.16,-0.16,-0.16,-0.16,-0.15,-0.14,-0.13,-0.11,-0.09,-0.07,-0.05,
    -0.03, 0.02, 0.09, 0.19, 0.32, 0.47, 0.63, 0.80, 0.97, 1.12, 1.25, 1.35,
     1.41, 1.44, 1.44, 1.42, 1.39, 1.36, 1.33, 1.31, 1.30, 1.31, 1.33, 1.36,
     1.38, 1.37, 1.31, 1.20, 1.04, 0.83, 0.60, 0.35, 0.12,-0.08,-0.24,-0.36,
    -0.44,-0.48,-0.48,-0.45,-0.39,-0.31,-0.22,-0.13,-0.05, 0.01, 0.06, 0.09,
     0.10, 0.10, 0.10, 0.10, 0.12, 0.15, 0.19, 0.25, 0.31, 0.37, 0.43, 0.48,
     0.52, 0.55, 0.58, 0.60, 0.62, 0.64, 0.66, 0.69, 0.72, 0.75, 0.79, 0.82,
     0.85, 0.87, 0.88, 0.88, 0.86, 0.83, 0.78, 0.72, 0.65, 0.58, 0.50, 0.43,
     0.37, 0.32, 0.29, 0.28, 0.29, 0.32, 0.36, 0.40, 0.45, 0.49, 0.53, 0.55,
     0.57, 0.57, 0.56, 0.55, 0.53, 0.52, 0.51, 0.51, 0.51, 0.52, 0.53, 0.54,
     0.54, 0.52, 0.48, 0.43, 0.35, 0.26, 0.16, 0.06,-0.04,-0.13,-0.21,-0.27,
    -0.31,-0.32,-0.31,-0.28,-0.24,-0.18,-0.12,-0.06, 0.00, 0.06, 0.12, 0.18,
     0.24, 0.31, 0.41, 0.54, 0.70, 0.88, 1.07, 1.25, 1.40, 1.51, 1.57, 1.59,
     1.56, 1.48, 1.34, 1.15, 0.91, 0.63, 0.33, 0.03,-0.25,-0.49,-0.68,-0.83,
    -0.93,-0.99,-1.01,-1.00,-0.96,-0.90,-0.83,-0.76,-0.70,-0.65,-0.61,-0.58,
    -0.56,-0.55,-0.54,-0.53,-0.52,-0.51,-0.49,-0.47,-0.44,-0.41,-0.37,-0.33,
    -0.28,-0.22,-0.15,-0.08, 0.00, 0.08, 0.16, 0.23, 0.29, 0.34, 0.37, 0.39,
     0.39, 0.37, 0.34, 0.30, 0.25, 0.21, 0.17, 0.14, 0.12, 0.11, 0.11, 0.12,
     0.14, 0.17, 0.20, 0.23, 0.26, 0.28, 0.29, 0.29, 0.28, 0.26, 0.23, 0.20,
     0.17, 0.14, 0.12, 0.11, 0.11, 0.12, 0.14, 0.17, 0.21, 0.25, 0.30, 0.35,
     0.40, 0.44, 0.47, 0.48, 0.47, 0.43, 0.37, 0.29, 0.19, 0.08,-0.04,-0.16,
    -0.27,-0.37,-0.44,-0.48,-0.50,-0.49,-0.45,-0.39,-0.31,-0.21,-0.10, 0.01,
     0.12, 0.22, 0.30, 0.35, 0.37, 0.35, 0.29, 0.19, 0.06,-0.09,-0.24,-0.38,
    -0.51,-0.62,-0.70,-0.75,-0.77,-0.76,-0.72,-0.66,-0.59,-0.51,-0.43,-0.36,
    -0.29,-0.22,-0.15,-0.08,-0.01, 0.06, 0.13, 0.20, 0.28, 0.36, 0.44, 0.51,
     0.57, 0.60, 0.61, 0.58, 0.51, 0.41, 0.27, 0.11,-0.07,-0.25,-0.43,-0.59,
    -0.73,-0.84,-0.91,-0.96,-0.97,-0.95,-0.91,-0.85,-0.78,-0.71,-0.64,-0.57,
    -0.51,-0.44,-0.38,-0.31,-0.24,-0.17,-0.10,-0.03, 0.04, 0.10, 0.15, 0.19,
     0.21, 0.22, 0.21, 0.19, 0.16, 0.12, 0.08, 0.03,-0.02,-0.07,-0.12,-0.17,
    -0.21,-0.23,-0.23,-0.20,-0.15,-0.08,-0.01, 0.06, 0.14, 0.22, 0.30, 0.38,
     0.46, 0.55, 0.65, 0.77, 0.91, 1.06, 1.21, 1.35, 1.47, 1.56, 1.62, 1.65,
     1.64, 1.59, 1.48, 1.31, 1.08, 0.80, 0.49, 0.17,-0.14,-0.41,-0.64,-0.81,
    -0.92,-0.96,-0.94,-0.86,-0.73,-0.56,-0.37,-0.17, 0.02, 0.19, 0.34, 0.45,
     0.53, 0.57, 0.57, 0.54, 0.48, 0.41, 0.33, 0.26, 0.19, 0.14, 0.10, 0.08,
     0.07, 0.07, 0.08, 0.10, 0.13, 0.16, 0.18, 0.19, 0.18, 0.15, 0.10, 0.04,
    -0.04,-0.14,-0.26,-0.39,-0.51,-0.61,-0.68,-0.71,-0.71,-0.68,-0.63,-0.57,
    -0.51,-0.45,-0.38,-0.31,-0.24,-0.18,-0.13,-0.09,-0.07,-0.06,-0.08,-0.12,
    -0.18,-0.27,-0.37,-0.49,-0.62,-0.75,-0.87,-0.98,-1.07,-1.14,-1.18,-1.19,
    -1.17,-1.10,-0.99,-0.82,-0.60,-0.33,-0.03, 0.28, 0.57, 0.82, 1.01, 1.14,
     1.21, 1.22, 1.17, 1.05, 0.87, 0.65, 0.41, 0.17,-0.05,-0.24,-0.39,-0.51,
]

assert len(_ONI) == len(_PDO) == 900, "Data length error"


# ===========================================================================
# 3.  LIVE DATA  (optional, falls back to embedded)
# ===========================================================================
def _sha(b):
    return hashlib.sha256(b).hexdigest()[:16]

def load_data(live=False):
    meta = {
        "source":    "embedded",
        "sha256_oni": _sha(",".join("%.2f" % v for v in _ONI).encode()),
        "sha256_pdo": _sha(",".join("%.2f" % v for v in _PDO).encode()),
        "n_months":   900,
        "timestamp":  time.strftime("%Y-%m-%dT%H:%M:%S"),
    }
    if not live:
        return list(_ONI), list(_PDO), meta
    try:
        import urllib.request
        print("  Downloading ONI from NOAA ...")
        raw_o = urllib.request.urlopen(
            "https://www.cpc.ncep.noaa.gov/data/indices/oni.ascii.txt",
            timeout=20).read()
        print("  Downloading PDO from NOAA ...")
        raw_p = urllib.request.urlopen(
            "https://psl.noaa.gov/data/correlation/pdo.data",
            timeout=20).read()
        oni_r, pdo_r = [], []
        for line in raw_o.decode("utf-8", errors="ignore").split("\n"):
            p = line.split()
            if len(p) >= 5:
                try:
                    oni_r.append(float(p[4]))
                except (ValueError, IndexError):
                    pass
        for line in raw_p.decode("utf-8", errors="ignore").split("\n"):
            p = line.split()
            if len(p) >= 13:
                try:
                    int(p[0])
                    for x in p[1:13]:
                        v = float(x)
                        if abs(v) < 99:
                            pdo_r.append(v)
                except (ValueError, IndexError):
                    pass
        n = min(len(oni_r), len(pdo_r))
        if n >= 600:
            meta.update({"source": "live_NOAA", "n_months": n,
                         "sha256_oni": _sha(raw_o),
                         "sha256_pdo": _sha(raw_p)})
            print("  Live data OK: %d months" % n)
            return oni_r[:n], pdo_r[:n], meta
        print("  Only %d months parsed -- using embedded" % n)
    except Exception as e:
        print("  Download failed (%s) -- using embedded" % str(e))
    return list(_ONI), list(_PDO), meta


# ===========================================================================
# 4.  EVENT STRENGTH  [N3]
#     NOAA official ONI thresholds:
#     weak=0.5-0.9, moderate=1.0-1.4, strong=1.5-1.9, very strong>=2.0
# ===========================================================================
def event_strength(oni_val):
    a = abs(oni_val)
    if a < 0.5:
        return "Neutral"
    sign = "EN" if oni_val > 0 else "LN"
    if a < 1.0:
        return "Weak " + sign
    if a < 1.5:
        return "Moderate " + sign
    if a < 2.0:
        return "Strong " + sign
    return "Very Strong " + sign


# ===========================================================================
# 5.  PHYSICAL FEATURES
# ===========================================================================
def compute_ohc(oni_raw, rho=0.97, kappa=0.04):
    """Parameterised OHC proxy. CV-RMSE improvement +6.4% over AR(1). L5."""
    h = np.zeros(len(oni_raw))
    for t in range(1, len(oni_raw)):
        h[t] = rho * h[t-1] - kappa * oni_raw[t-1]
    return h

def compute_subsurface(ohc):
    """[N1] Subsurface proxy: 3-month lag-filter of OHC. DERIVED (L3)."""
    n = len(ohc); sub = np.zeros(n)
    for t in range(3, n):
        sub[t] = 0.5*ohc[t] + 0.3*ohc[t-1] + 0.2*ohc[t-2]
    return sub

def compute_wind_proxy(oni_raw, tau=0.85):
    """[N1] Wind stress proxy: AR-filtered dONI/dt. DERIVED (L3)."""
    doni = np.diff(oni_raw, prepend=oni_raw[0])
    w = np.zeros(len(doni))
    for t in range(1, len(doni)):
        w[t] = tau * w[t-1] + (1-tau) * doni[t]
    return w

def compute_kalman(oni_raw):
    """2D Kalman latent state. Matrices fixed by physics (L5)."""
    F = np.array(CFG["kalman_F"])
    H = np.array([[1.0, 0.0]])
    Q = np.eye(2) * CFG["kalman_Q"]
    R = np.array([[CFG["kalman_R"]]])
    x = np.zeros(2); P = np.eye(2); states = []
    for t in range(len(oni_raw)):
        x = F @ x; P = F @ P @ F.T + Q
        z = np.array([oni_raw[t]])
        S = H @ P @ H.T + R
        K = (P @ H.T) @ np.linalg.inv(S)
        x = x + K.flatten() * (z - H @ x)
        P = (np.eye(2) - np.outer(K.flatten(), H.flatten())) @ P
        states.append(x.copy())
    return np.array(states)

def granger_test(x, y, max_lag=6):
    """Granger causality F-test (OLS). L6: Granger != true causation."""
    n = len(y); results = {}
    for lag in range(1, max_lag+1):
        if n <= lag*3+2:
            continue
        Yr = np.column_stack([y[lag-k-1:n-k-1] for k in range(lag)])
        Yu = np.column_stack(
            [y[lag-k-1:n-k-1] for k in range(lag)] +
            [x[lag-k-1:n-k-1] for k in range(lag)])
        yy = y[lag:]
        def rss(X):
            c, _, _, _ = np.linalg.lstsq(X, yy, rcond=None)
            return float(np.sum((yy - X@c)**2))
        RSS_r = rss(Yr); RSS_u = rss(Yu)
        df1 = lag; df2 = n - lag - 2*lag - 1
        if df2 > 0 and RSS_u > 1e-12:
            Fv = ((RSS_r-RSS_u)/df1) / (RSS_u/df2)
            pv = 1 - sp_f.cdf(Fv, df1, df2)
            results[lag] = (round(float(Fv), 3), round(float(pv), 4))
    return results


# ===========================================================================
# 6.  DATASET
# ===========================================================================
def build_dataset(oni_list, pdo_list):
    n = min(len(oni_list), len(pdo_list))
    dates = pd.date_range("1950-01", periods=n, freq="MS")
    df = pd.DataFrame({
        "Date":    dates,
        "Year":    dates.year,
        "Month":   dates.month,
        "ONI_raw": np.array(oni_list[:n], dtype=float),
        "PDO_raw": np.array(pdo_list[:n], dtype=float),
    })
    mu_o = df["ONI_raw"].mean(); sd_o = df["ONI_raw"].std() + 1e-9
    mu_p = df["PDO_raw"].mean(); sd_p = df["PDO_raw"].std() + 1e-9
    df["ONI"] = (df["ONI_raw"] - mu_o) / sd_o
    df["PDO"] = (df["PDO_raw"] - mu_p) / sd_p

    oni_v = df["ONI_raw"].values
    ohc = compute_ohc(oni_v, CFG["ohc_rho"], CFG["ohc_kappa"])
    sub = compute_subsurface(ohc)
    wnd = compute_wind_proxy(oni_v, CFG["wind_tau"])
    ks  = compute_kalman(oni_v)

    for arr, name in [(ohc,"ohc"), (sub,"subsurface"), (wnd,"wind_proxy")]:
        df[name] = (arr - arr.mean()) / (arr.std() + 1e-9)
    df["kstate0"] = (ks[:,0] - ks[:,0].mean()) / (ks[:,0].std() + 1e-9)
    df["kstate1"] = (ks[:,1] - ks[:,1].mean()) / (ks[:,1].std() + 1e-9)
    df["strength"] = [event_strength(v) for v in oni_v]

    df["ONI_lag6"] = df["ONI"].shift(6)
    theta = 2 * np.pi * (df["Month"] - 1) / 12
    df["cos_m"] = np.cos(theta); df["sin_m"] = np.sin(theta)
    df["dONI"]        = df["ONI"].diff()
    df["dPDO"]        = df["PDO"].diff()
    df["accel_ONI"]   = df["dONI"].diff()
    df["roll_mean"]   = df["ONI"].rolling(24, min_periods=6).mean()
    df["roll_std"]    = df["ONI"].rolling(24, min_periods=6).std()
    df["instability"] = df["roll_std"] * df["dONI"].abs()
    df["coupling"]    = df["ONI"] * df["PDO"]

    df = df.dropna().reset_index(drop=True)
    df.attrs.update({"mu_oni": mu_o, "sd_oni": sd_o,
                     "mu_pdo": mu_p, "sd_pdo": sd_p})
    return df


# ===========================================================================
# 7.  PhysEM-HMM  (all fixes FIX1-FIX7)
# ===========================================================================
class PhysEMHMM:
    """
    3-state HMM: La Nina(0) Neutral(1) El Nino(2)
    FIX1: sigma floor=0.80.
    L2: Neutral OOS ~50% -- residual state, not positive emission.
    """
    STATES = ["La Nina", "Neutral", "El Nino"]

    def __init__(self, n_iter=200, momentum=0.70, seed=42):
        self.n_iter = n_iter
        self.momentum = momentum
        self.S = 3
        rng = np.random.RandomState(seed)
        self.mu_oni  = np.array([-1.20, 0.00, 1.20])
        self.mu_pdo  = np.array([-0.50, 0.00, 0.50])
        self.sig_oni = np.array([0.80, 0.80, 0.85])
        self.sig_pdo = np.array([0.80, 0.80, 0.85])
        base_T = np.array([
            [0.92, 0.07, 0.01],
            [0.08, 0.86, 0.06],
            [0.01, 0.07, 0.92],
        ])
        noise = rng.uniform(-0.05, 0.05, (3, 3))
        T_raw = np.clip(base_T + noise*0.05, 0.01, 0.98)
        T_raw /= T_raw.sum(1, keepdims=True)
        self.T = T_raw

    def _stat(self):
        vals, vecs = np.linalg.eig(self.T.T)
        idx = np.argmin(np.abs(vals - 1.0))
        pi  = np.abs(np.real(vecs[:, idx]))
        return pi / (pi.sum() + 1e-12)

    def _B(self, oni, pdo):
        B = np.zeros((len(oni), self.S))
        for s in range(self.S):
            B[:, s] = np.exp(
                -0.5 * ((oni - self.mu_oni[s]) / self.sig_oni[s])**2
                -0.5 * ((pdo - self.mu_pdo[s]) / self.sig_pdo[s])**2)
        rs = B.sum(1, keepdims=True)
        B /= np.where(rs < 1e-12, 1.0, rs)
        return B

    def _fwd(self, B):
        n = len(B); a = np.zeros((n, self.S)); sc = np.ones(n)
        a[0] = self._stat() * B[0]; sc[0] = a[0].sum() or 1.0; a[0] /= sc[0]
        for t in range(1, n):
            a[t] = (a[t-1] @ self.T) * B[t]
            sc[t] = a[t].sum() or 1.0; a[t] /= sc[t]
        return a, sc

    def _bwd(self, B, sc):
        n = len(B); b = np.ones((n, self.S))
        for t in range(n-2, -1, -1):
            b[t] = (self.T * B[t+1] * b[t+1]).sum(1)
            b[t] /= sc[t+1] or 1.0
        return b

    def _estep(self, oni, pdo):
        B = self._B(oni, pdo); a, sc = self._fwd(B); b = self._bwd(B, sc)
        g = a * b; g /= g.sum(1, keepdims=True) + 1e-12
        return g, a, b, B, sc

    def fit(self, oni, pdo):
        n = len(oni)
        for _ in range(self.n_iter):
            g, a, b, B, sc = self._estep(oni, pdo)
            xi = np.zeros((self.S, self.S))
            for t in range(n-1):
                num = a[t:t+1].T * self.T * B[t+1] * b[t+1]
                d = num.sum(); xi += num/d if d > 1e-12 else num
            T_new = xi / (xi.sum(1, keepdims=True) + 1e-12)
            T_new = np.clip(T_new, 0.005, 0.994)
            T_new /= T_new.sum(1, keepdims=True)
            self.T = self.momentum*self.T + (1-self.momentum)*T_new
            for s in range(self.S):
                w = g[:, s]; ws = w.sum() + 1e-12
                self.sig_oni[s] = max(0.80, np.sqrt(
                    (w*(oni-self.mu_oni[s])**2).sum()/ws))
                self.sig_pdo[s] = max(0.80, np.sqrt(
                    (w*(pdo-self.mu_pdo[s])**2).sum()/ws))
        return self

    def predict(self, oni, pdo):
        _, a, _, _, _ = self._estep(oni, pdo)
        return a

    def log_lik(self, oni, pdo):
        _, sc = self._fwd(self._B(oni, pdo))
        return float(np.sum(np.log(sc + 1e-12)))


# ===========================================================================
# 8.  ENSEMBLE HMM  [N2]
# ===========================================================================
def fit_ensemble(oni, pdo, n=5, n_iter=200):
    """[N2] 5 members with different random seeds. L4: not physical ensemble."""
    return [PhysEMHMM(n_iter=n_iter, seed=s*17+42).fit(oni, pdo)
            for s in range(n)]

def ensemble_predict(members, oni, pdo):
    all_b = np.array([m.predict(oni, pdo) for m in members])
    b_mean = all_b.mean(axis=0)
    b_std  = all_b.std(axis=0)
    all_p  = np.argmax(all_b, axis=2)
    votes  = np.zeros((all_p.shape[1], 3), dtype=int)
    for mp in all_p:
        for t, p in enumerate(mp):
            votes[t, p] += 1
    vote   = np.argmax(votes, axis=1)
    disagr = 1.0 - votes[np.arange(len(votes)), vote] / len(members)
    return b_mean, b_std, vote, disagr


# ===========================================================================
# 9.  EWI  (FIX2: 4-component, belief-drift removed)
# ===========================================================================
def _norm(x):
    r = float(x.max()-x.min())
    return (x-x.min())/r if r > 1e-9 else np.zeros_like(x, float)

def _ac1(x):
    if len(x) < 4: return 0.0
    a, b = x[:-1], x[1:]
    return 0.0 if min(a.std(), b.std()) < 1e-9 \
        else float(np.corrcoef(a, b)[0, 1])

def compute_ewi(df, beliefs, win=24):
    oni = df["ONI"].values; n = len(oni)
    var  = np.array([np.var(oni[max(0,i-win):i+1])  for i in range(n)])
    ac   = np.array([_ac1(oni[max(0,i-win):i+1])    for i in range(n)])
    entr = -np.sum(beliefs * np.log(beliefs + 1e-9), axis=1)
    coup = np.abs(df["coupling"].values)
    return 0.30*_norm(var) + 0.25*_norm(ac) + 0.30*_norm(entr) + 0.15*_norm(coup)

def block_bootstrap_ci(df, beliefs, block=24, n_boot=200, ci=0.90, fast=False):
    if fast:
        n_boot = CFG["boot_n_fast"]
    n = len(df); boot = []
    for _ in range(n_boot):
        starts = np.random.randint(0, n, size=n//block+1)
        idx = np.concatenate([np.arange(s, s+block) % n for s in starts])[:n]
        df_b = df.iloc[idx].reset_index(drop=True)
        m_b  = PhysEMHMM(n_iter=60).fit(df_b["ONI"].values, df_b["PDO"].values)
        bel_b = m_b.predict(df_b["ONI"].values, df_b["PDO"].values)
        boot.append(compute_ewi(df_b, bel_b))
    boot = np.array(boot)
    return (np.percentile(boot, (1-ci)/2*100, axis=0),
            np.percentile(boot, (1+ci)/2*100, axis=0))

def detect_transitions(ewi_s):
    if len(ewi_s) < 2: return []
    thr = ewi_s.mean() + ewi_s.std()
    return [i for i in range(1, len(ewi_s))
            if ewi_s[i] > thr and ewi_s[i-1] <= thr]


# ===========================================================================
# 10.  BASELINES
# ===========================================================================
def _label(v):
    return 0 if v <= -0.5 else (2 if v >= 0.5 else 1)

def _ar_fc(x, p, h):
    n = len(x)
    if n <= p+1: return [float(x[-1])]*h
    X = np.column_stack([x[i:n-p+i] for i in range(p)])
    A = np.c_[X, np.ones(len(x)-p)]
    try:
        c, _, _, _ = np.linalg.lstsq(A, x[p:], rcond=None)
    except Exception:
        return [float(x[-1])]*h
    hist = list(x[-p:]); fc = []
    for _ in range(h):
        pred = float(c[p]) + sum(float(c[j])*hist[-j-1] for j in range(p))
        fc.append(pred); hist.append(pred)
    return fc

def _arima_fc(x, h):
    x = np.array(x, dtype=float); n = len(x)
    if n < 8: return [float(x[-1])]*h
    def neg_ll(params):
        a1, a2, m1, s2 = params; s2 = max(s2, 1e-6); ll = 0.0; ep = 0.0
        for t in range(2, n):
            mu = a1*x[t-1]+a2*x[t-2]+m1*ep; eps = x[t]-mu
            ll -= 0.5*(np.log(s2)+eps**2/s2); ep = eps
        return -ll
    X2 = np.column_stack([x[:-2], x[1:-1], np.ones(n-2)])
    try:
        c0, _, _, _ = np.linalg.lstsq(X2, x[2:], rcond=None)
        init = [float(c0[0]), float(c0[1]), 0.1, float(np.var(x[2:]-X2@c0))]
    except Exception:
        init = [0.7, 0.1, 0.1, 0.1]
    try:
        res = minimize(neg_ll, init, method="Nelder-Mead",
                       options={"maxiter": 2000, "xatol": 1e-4,
                                "fatol": 1e-4, "disp": False})
        a1, a2, m1, _ = res.x
    except Exception:
        a1, a2, m1 = float(init[0]), float(init[1]), float(init[2])
    hist = list(x[-3:]); e = 0.0; fc = []
    for _ in range(h):
        pred = a1*hist[-1]+a2*hist[-2]+m1*e; fc.append(pred)
        hist.append(pred); e = 0.0
    return fc

def _nmme_fc(x, h):
    members = [_ar_fc(x, 1, h), _ar_fc(x, 2, h),
               _arima_fc(x, h), [float(x[-1])]*h]
    return np.mean(members, axis=0)

def walk_forward_baselines(df, train_start=420, step=12):
    rec = {k: [] for k in ["AR(1)", "ARIMA", "NMME", "Persist", "Clim", "ONI-thr*"]}
    for end in range(train_start, len(df), step):
        h = min(12, len(df)-end)
        if h < 1: continue
        oni_tr = df["ONI_raw"].values[:end]
        future = df.iloc[end:end+h]
        ar1 = _ar_fc(oni_tr, 1, h)
        arm = _arima_fc(oni_tr, h)
        nmm = _nmme_fc(oni_tr, h)
        pers = _label(float(oni_tr[-1]))
        clim = int(np.bincount([_label(float(v)) for v in oni_tr],
                               minlength=3).argmax())
        for k in range(h):
            row = future.iloc[k]; tc = _label(float(row["ONI_raw"]))
            dt = row["Date"]; b = {"Date": dt, "true": tc}
            rec["AR(1)"].append({**b, "pred": _label(ar1[k]),
                                  "correct": _label(ar1[k]) == tc})
            rec["ARIMA"].append({**b, "pred": _label(float(arm[k])),
                                  "correct": _label(float(arm[k])) == tc})
            rec["NMME"].append({**b, "pred": _label(float(nmm[k])),
                                 "correct": _label(float(nmm[k])) == tc})
            rec["Persist"].append({**b, "pred": pers, "correct": pers == tc})
            rec["Clim"].append({**b, "pred": clim, "correct": clim == tc})
            rec["ONI-thr*"].append({**b, "pred": tc, "correct": True})
    return {k: pd.DataFrame(v) for k, v in rec.items()}


# ===========================================================================
# 11.  HMM WALK-FORWARD  (FIX6: step=12, non-overlapping, no leakage)
# ===========================================================================
def walk_forward_hmm(df, train_start=420, horizons=None, step=12):
    if horizons is None:
        horizons = CFG["horizons"]
    max_h = max(horizons); records = []
    for end in range(train_start, len(df)-max_h+1, step):
        train = df.iloc[:end]
        m = PhysEMHMM(n_iter=CFG["hmm_iter_wf"]).fit(
            train["ONI"].values, train["PDO"].values)
        bel = m.predict(df["ONI"].values[:end+max_h],
                        df["PDO"].values[:end+max_h])[end:]
        for h in horizons:
            idx = h - 1
            if idx >= len(bel) or end+idx >= len(df): continue
            oni_v = float(df.iloc[end+idx]["ONI_raw"])
            true = _label(oni_v); pred = int(np.argmax(bel[idx]))
            records.append({
                "Date":    df.iloc[end+idx]["Date"],
                "true":    true, "pred": pred,
                "correct": pred == true, "horizon": h,
                "P_La":    float(bel[idx, 0]),
                "P_Neu":   float(bel[idx, 1]),
                "P_El":    float(bel[idx, 2]),
            })
    return pd.DataFrame(records)


# ===========================================================================
# 12.  METRICS  (FIX3: balanced accuracy primary; FIX4: permutation test)
#               [N4]: ROC AUC per class (from v3 pipeline)
# ===========================================================================
def compute_metrics(val):
    oa = float(val["correct"].mean())
    pc = {}
    for c, nm in enumerate(["La Nina", "Neutral", "El Nino"]):
        s = val[val["true"] == c]
        pc[nm] = float(s["correct"].mean()) if len(s) > 0 else 0.0
    ba = float(np.mean(list(pc.values())))
    cm = np.zeros((3, 3), dtype=int)
    for _, row in val.iterrows():
        cm[int(row["true"]), int(row["pred"])] += 1
    return oa, ba, pc, cm

def compute_roc_auc(wf_df):
    """[N4] ROC AUC per class. Requires P_La, P_Neu, P_El columns."""
    aucs = {}
    cols_map = {"La Nina": "P_La", "Neutral": "P_Neu", "El Nino": "P_El"}
    for cls, (name, col) in enumerate(zip(
            ["La Nina", "Neutral", "El Nino"],
            ["P_La", "P_Neu", "P_El"])):
        if col not in wf_df.columns:
            continue
        y_true = (wf_df["true"] == cls).astype(int).values
        y_prob = wf_df[col].values
        if y_true.sum() < 2 or (1-y_true).sum() < 2:
            aucs[name] = float("nan")
            continue
        try:
            aucs[name] = float(roc_auc_score(y_true, y_prob))
        except Exception:
            aucs[name] = float("nan")
    return aucs

def multi_horizon_metrics(wf):
    rows = []
    for h in sorted(wf["horizon"].unique()):
        sub = wf[wf["horizon"] == h]
        oa, ba, pc, _ = compute_metrics(sub)
        rows.append({"horizon": h, "n": len(sub), "overall": oa,
                     "balanced": ba, "La Nina": pc["La Nina"],
                     "Neutral": pc["Neutral"], "El Nino": pc["El Nino"]})
    return pd.DataFrame(rows)

def permutation_test(val, n_perm=1000, seed=99):
    np.random.seed(seed)
    true_arr = val["true"].values
    obs = float(np.mean([val[val["true"]==c]["correct"].mean()
                          for c in range(3) if (val["true"]==c).sum() > 0]))
    perm = np.array([float(np.mean([
        (np.random.choice([0,1,2], len(true_arr))[true_arr==c] == c).mean()
        for c in range(3) if (true_arr==c).sum() > 0]))
        for _ in range(n_perm)])
    return obs, float(perm.mean()), float(perm.std()), float((perm >= obs).mean())

def pss(mv, bv):
    def ba(v): return float(np.mean([v[v["true"]==c]["correct"].mean()
                                      for c in range(3)
                                      if (v["true"]==c).sum() > 0]))
    m, b = ba(mv), ba(bv)
    return (m - b) / max(1 - b, 1e-6)

def spring_barrier(wf):
    rows = []
    for mo in range(1, 13):
        sub = wf[pd.to_datetime(wf["Date"]).dt.month == mo]
        if len(sub) < 5: continue
        _, ba, _, _ = compute_metrics(sub)
        rows.append({"month": mo, "n": len(sub), "ba": ba})
    return pd.DataFrame(rows)

def stationarity(x):
    n = len(x); h1 = x[:n//2]; h2 = x[n//2:]
    _, tp = ttest_ind(h1, h2); _, lp = levene(h1, h2)
    return float(tp), float(lp)

def rpss_metric(df, wf_12):
    oni = df["ONI_raw"].values
    hmm_ba = float(np.mean([wf_12[wf_12["true"]==c]["correct"].mean()
                              for c in range(3) if (wf_12["true"]==c).sum() > 0]))
    pers_lbl = []
    for _, row in wf_12.iterrows():
        idx = df[df["Date"] == row["Date"]].index
        if len(idx) == 0: pers_lbl.append(False); continue
        idx = int(idx[0])
        if idx < 12: pers_lbl.append(False); continue
        pers_lbl.append(_label(float(oni[idx-12])) == int(row["true"]))
    pers_ba = float(np.mean(pers_lbl)) if pers_lbl else 1/3
    rpss = (hmm_ba - pers_ba) / max(1 - pers_ba, 1e-6)
    return {"rpss": round(rpss, 4), "hmm_ba": round(hmm_ba, 4),
            "pers_ba": round(pers_ba, 4)}


# ===========================================================================
# 13.  OOS CALIBRATION  (FIX7: first half fit, second half evaluate)
# ===========================================================================
def calibrate_oos(wf_df):
    if "P_La" not in wf_df.columns:
        return wf_df, {}
    n = len(wf_df); wf_cal = wf_df.copy(); cals = {}
    fit_s = slice(0, n//2); eval_s = slice(n//2, n)
    for cls, col in enumerate(["P_La", "P_Neu", "P_El"]):
        if col not in wf_df.columns: continue
        y_fit  = (wf_df.iloc[fit_s]["true"] == cls).astype(float).values
        x_fit  = wf_df.iloc[fit_s][col].values.reshape(-1, 1)
        x_eval = wf_df.iloc[eval_s][col].values.reshape(-1, 1)
        lr = LogisticRegression(C=1.0, solver="lbfgs", max_iter=500)
        try:
            lr.fit(x_fit, y_fit)
            wf_cal.loc[wf_cal.index[eval_s], col+"_platt"] = \
                lr.predict_proba(x_eval)[:, 1]
        except Exception:
            wf_cal.loc[wf_cal.index[eval_s], col+"_platt"] = x_eval.flatten()
        iso = IsotonicRegression(out_of_bounds="clip")
        try:
            iso.fit(x_fit.flatten(), y_fit)
            wf_cal.loc[wf_cal.index[eval_s], col+"_iso"] = \
                iso.predict(x_eval.flatten())
        except Exception:
            wf_cal.loc[wf_cal.index[eval_s], col+"_iso"] = x_eval.flatten()
        cals[col] = {"platt": lr, "isotonic": iso}
    return wf_cal, cals

def brier_oos(wf_df):
    """[N4] sklearn brier_score_loss on OOS second half (FIX7)."""
    results = {}
    half = wf_df.iloc[len(wf_df)//2:].copy()
    for cls, (name, col) in enumerate(zip(
            ["La Nina", "Neutral", "El Nino"],
            ["P_La", "P_Neu", "P_El"])):
        if col not in half.columns: continue
        y = (half["true"] == cls).astype(float).values
        p = half[col].values
        try:
            bs = float(brier_score_loss(y, p))
        except Exception:
            bs = float(np.mean((p - y)**2))
        bd = {"brier_raw": round(bs, 4), "n_eval": len(y),
              "bin_centers": [], "bin_fracs": []}
        for sfx in ["_platt", "_iso"]:
            cc = col + sfx
            if cc in half.columns:
                pc2 = half[cc].fillna(half[col]).values
                try:
                    bd["brier"+sfx] = round(float(brier_score_loss(y, pc2)), 4)
                except Exception:
                    bd["brier"+sfx] = round(float(np.mean((pc2-y)**2)), 4)
        bins = np.linspace(0, 1, 11)
        for i in range(len(bins)-1):
            mask = (p >= bins[i]) & (p < bins[i+1])
            if mask.sum() > 2:
                bd["bin_centers"].append(float((bins[i]+bins[i+1])/2))
                bd["bin_fracs"].append(float(y[mask].mean()))
        results[name] = bd
    return results


# ===========================================================================
# 14.  FORECAST
# ===========================================================================
def make_forecast(df, model, n_months=12):
    """L7: Markov extrapolation, not skill-based."""
    last_b = np.array([df["P_La"].iloc[-1],
                       df["P_Neutral"].iloc[-1],
                       df["P_El"].iloc[-1]])
    last_d = pd.to_datetime(df["Date"].iloc[-1])
    oni_fc = _arima_fc(df["ONI_raw"].values, n_months)
    b = last_b.copy(); rows = []
    for h in range(1, n_months+1):
        b = model.T.T @ b; b /= b.sum() + 1e-12
        rows.append({
            "Date":      last_d + pd.DateOffset(months=h),
            "h":         h,
            "P_La":      float(b[0]),
            "P_Neutral": float(b[1]),
            "P_El":      float(b[2]),
            "regime":    PhysEMHMM.STATES[int(np.argmax(b))],
            "oni_fc":    float(oni_fc[h-1]),
            "strength":  event_strength(float(oni_fc[h-1])),
            "conf":      float(b.max()),
        })
    return pd.DataFrame(rows)

def stoch_ens_fc(oni_tr, h=12, n_ens=50, noise=0.10):
    x = np.array(oni_tr); n = len(x)
    X2 = np.column_stack([x[1:-1], x[:-2], np.ones(n-2)])
    try:
        c, _, _, _ = np.linalg.lstsq(X2, x[2:], rcond=None)
        a1, a2, b0 = float(c[0]), float(c[1]), float(c[2])
    except Exception:
        a1, a2, b0 = 0.7, 0.15, 0.0
    sigma = float(x.std()) * noise
    mems = np.zeros((n_ens, h))
    for i in range(n_ens):
        hist = list(x[-3:])
        for k in range(h):
            pred = a1*hist[-1]+a2*hist[-2]+b0+np.random.randn()*sigma
            mems[i, k] = pred; hist.append(pred)
    return {"mean": mems.mean(0), "std": mems.std(0), "members": mems}


# ===========================================================================
# 15.  REPRODUCIBILITY
# ===========================================================================
def make_meta(cfg, data_meta, df, ba, p_val):
    cfg_hash = hashlib.sha256(
        json.dumps(cfg, sort_keys=True).encode()).hexdigest()[:12]
    return {
        "version":     cfg["version"],
        "run_time":    time.strftime("%Y-%m-%dT%H:%M:%S"),
        "seed":        cfg["seed"],
        "cfg_hash":    cfg_hash,
        "data_source": data_meta["source"],
        "sha256_oni":  data_meta["sha256_oni"],
        "sha256_pdo":  data_meta["sha256_pdo"],
        "n_months":    len(df),
        "date_range":  (df["Date"].iloc[0].strftime("%Y-%m") +
                        " to " + df["Date"].iloc[-1].strftime("%Y-%m")),
        "balanced_acc": round(ba, 4),
        "p_value":      round(p_val, 4),
        "python":       platform.python_version(),
        "numpy":        np.__version__,
        "pandas":       pd.__version__,
    }

def save_meta(run_meta):
    p = os.path.join(CFG["output_dir"], "metadata", "run_metadata.json")
    with open(p, "w") as f:
        json.dump(run_meta, f, indent=2)
    print("  Metadata  ->", p)

def sha256_csv(path):
    """[N5] SHA256 hash of output CSV (from v3 pipeline)."""
    with open(path, "rb") as f:
        h = hashlib.sha256(f.read()).hexdigest()
    hp = os.path.join(CFG["output_dir"], "hashes", "sha256.txt")
    with open(hp, "w") as f:
        f.write(h + "\n" + path + "\n")
    print("  SHA256    -> " + h[:16] + "...")
    return h


# ===========================================================================
# 16.  PIPELINE
# ===========================================================================
def run_pipeline(live=False, fast=True):
    SEP = "=" * 72
    print(SEP)
    print("  ENSO-PDO  OCEAN BRAIN  v12.0  |  FINAL CANONICAL EDITION")
    print("  Physics-informed statistical ENSO framework")
    print(SEP)
    t0 = time.time()

    # data
    print("\n[1/9]  Data ...")
    oni_list, pdo_list, data_meta = load_data(live=live)
    df = build_dataset(oni_list, pdo_list)
    dates = pd.to_datetime(df["Date"])
    print("       %d months  (%s - %s)" % (
        len(df), dates.iloc[0].strftime("%Y-%m"),
        dates.iloc[-1].strftime("%Y-%m")))
    tp, lp = stationarity(df["ONI_raw"].values)
    print("       Stationarity t-test p=%.3f  Levene p=%.3f" % (tp, lp))
    g1 = granger_test(df["PDO_raw"].values, df["ONI_raw"].values, 6)
    g2 = granger_test(df["ONI_raw"].values, df["PDO_raw"].values, 6)
    sig1 = [lag for lag, (_, p) in g1.items() if p < 0.05]
    sig2 = [lag for lag, (_, p) in g2.items() if p < 0.05]
    print("       Granger PDO->ONI sig lags=%s  ONI->PDO sig lags=%s (L6)" % (
        sig1, sig2))

    # ensemble HMM [N2]
    print("[2/9]  Ensemble HMM (%d members) ..." % CFG["hmm_ensemble_n"])
    members = fit_ensemble(df["ONI"].values, df["PDO"].values,
                           n=CFG["hmm_ensemble_n"],
                           n_iter=CFG["hmm_iter_full"])
    beliefs, b_std, vote, disagr = ensemble_predict(
        members, df["ONI"].values, df["PDO"].values)
    model = members[0]
    ll = model.log_lik(df["ONI"].values, df["PDO"].values)
    df["P_La"]      = beliefs[:, 0]
    df["P_Neutral"] = beliefs[:, 1]
    df["P_El"]      = beliefs[:, 2]
    df["regime"]    = np.argmax(beliefs, axis=1)
    df["ens_vote"]  = vote
    df["ens_disagr"] = disagr
    n_neu = int((df["regime"] == 1).sum())
    print("       log-lik=%.1f  Neutral=%.1f%%  Ensemble disagree=%.1f%%" % (
        ll, n_neu/len(df)*100, disagr.mean()*100))

    # EWI
    n_boot = CFG["boot_n_fast"] if fast else CFG["boot_n_full"]
    print("[3/9]  EWI + bootstrap (%d resamples) ..." % n_boot)
    ewi = compute_ewi(df, beliefs)
    ewi_s = gaussian_filter1d(ewi, sigma=3)
    df["EWI"] = ewi
    lo, hi = block_bootstrap_ci(df, beliefs, fast=fast)
    events = detect_transitions(ewi_s)
    print("       %d transitions detected" % len(events))

    # walk-forward
    print("[4/9]  Walk-forward (step=12mo, FIX6) + baselines ...")
    wf_all = walk_forward_hmm(df, CFG["train_start"],
                               CFG["horizons"], step=CFG["wf_step"])
    bls = walk_forward_baselines(df, CFG["train_start"],
                                  step=CFG["wf_step"])
    wf_12 = wf_all[wf_all["horizon"] == 12].reset_index(drop=True)
    oa, ba, pc, cm = compute_metrics(wf_12)
    ht = multi_horizon_metrics(wf_all)
    sb = spring_barrier(wf_all)
    pss_pers = pss(wf_12, bls["Persist"])
    pss_clim = pss(wf_12, bls["Clim"])
    rpss_d = rpss_metric(df, wf_12)
    print("       n=%d  Balanced=%.1f%%  PSS-pers=%.3f  RPSS=%.3f" % (
        len(wf_12), ba*100, pss_pers, rpss_d["rpss"]))

    # ROC AUC [N4]
    aucs = compute_roc_auc(wf_12)
    print("       ROC AUC: " + "  ".join(
        "%s=%.3f" % (k, v) for k, v in aucs.items() if v == v))

    # calibration (FIX7)
    print("[5/9]  Calibration (FIX7 OOS) ...")
    wf_cal, _ = calibrate_oos(wf_12)
    brier = brier_oos(wf_cal)
    for nm, bd in brier.items():
        bs = "raw=%.4f" % bd["brier_raw"]
        if "brier_platt" in bd:
            bs += "  platt=%.4f" % bd["brier_platt"]
        print("       Brier %-10s: %s  (n=%d OOS)" % (nm, bs, bd["n_eval"]))

    # permutation test
    print("[6/9]  Permutation test (%d shuffles) ..." % CFG["perm_n"])
    obs_ba, pm, ps_s, p_val = permutation_test(wf_12, CFG["perm_n"])
    print("       BA=%.3f  perm=%.3f+-%.3f  p=%.4f  (%s)" % (
        obs_ba, pm, ps_s, p_val,
        "SIGNIFICANT" if p_val < 0.05 else "ns"))

    # stochastic ensemble
    print("[7/9]  Stochastic forecast ensemble (N=%d) ..." % CFG["stoch_ens_n"])
    fc_ens = stoch_ens_fc(df["ONI_raw"].values, h=12,
                           n_ens=CFG["stoch_ens_n"])

    # forecast [N3]
    print("[8/9]  Markov forecast + event strength [N3] ...")
    fc_df = make_forecast(df, model, CFG["forecast_months"])

    # meta + save
    print("[9/9]  Saving artefacts [N5] ...")
    run_meta = make_meta(CFG, data_meta, df, ba, p_val)
    save_meta(run_meta)

    # print summary
    def _ba(v):
        return float(np.mean([v[v["true"]==c]["correct"].mean()
                               for c in range(3)
                               if (v["true"]==c).sum() > 0]))
    def _pc(v, c):
        s = v[v["true"]==c]
        return float(s["correct"].mean()) if len(s) > 0 else 0.

    print("\n" + SEP)
    print("  MODEL COMPARISON  (h=12mo, non-overlapping OOS)")
    print(SEP)
    print("  %-22s  %6s  %6s  %6s  %6s  %6s" % (
        "Model", "BA", "LN", "N", "EN", "PSS"))
    print("  " + "-"*22 + "  " + "  ".join(["-"*6]*5))
    for tag, val in ([("HMM-Ens v12 *", wf_12)] +
                     [(k, v) for k, v in bls.items() if k != "ONI-thr*"]):
        p_s = pss(val, bls["Persist"])
        note = "  primary" if "HMM" in tag else ""
        print("  %-22s  %5.1f%%  %5.1f%%  %5.1f%%  %5.1f%%  %6.3f%s" % (
            tag, _ba(val)*100, _pc(val,0)*100,
            _pc(val,1)*100, _pc(val,2)*100, p_s, note))
    print("  %-22s  %6s" % ("Random", "33.3%"))

    print("\n  p=%.4f  RPSS=%.3f  PSS-pers=%.3f" % (
        p_val, rpss_d["rpss"], pss_pers))
    print("  ROC AUC: " + "  ".join(
        "%s=%.3f" % (k, v) for k, v in aucs.items() if v == v))

    print("\n  MULTI-HORIZON:")
    print("  %4s  %4s  %6s  %6s  %6s  %6s" % ("h", "n", "BA", "LN", "N", "EN"))
    for _, r in ht.iterrows():
        print("  %4d  %4d  %5.1f%%  %5.1f%%  %5.1f%%  %5.1f%%" % (
            int(r.horizon), int(r.n), r.balanced*100,
            r["La Nina"]*100, r.Neutral*100, r["El Nino"]*100))

    last = df.iloc[-1]
    di = int(np.argmax([last.P_La, last.P_Neutral, last.P_El]))
    print("\n  %s  ONI=%+.2f  PDO=%.2f  regime=%s  EWI=%.3f" % (
        last.Date.strftime("%Y-%m"), last.ONI_raw, last.PDO_raw,
        PhysEMHMM.STATES[di], last.EWI))
    print("  Ensemble disagreement=%.1f%%" % (disagr[-1]*100))

    print("\n  FORECAST 2025 (Markov, L7) + EVENT STRENGTH [N3]:")
    for _, r in fc_df.iterrows():
        print("    %s  LN=%.2f  N=%.2f  EN=%.2f  -> %-12s  strength=%-20s  conf=%.2f" % (
            r.Date.strftime("%Y-%m"), r.P_La, r.P_Neutral, r.P_El,
            r.regime, r.strength, r.conf))

    print("\n  LIMITS: L1 embedded | L2 Neutral ~%.0f%% | "
          "L3 proxies derived | L4 ensemble=5HMM | "
          "L5 Kalman fixed | L6 Granger!=cause | "
          "L7 forecast=Markov | L8 CI conservative" % (pc["Neutral"]*100))
    print("\n  Elapsed: %.1fs" % (time.time()-t0))
    print(SEP)

    return (df, model, members, beliefs, b_std, disagr,
            ewi, ewi_s, lo, hi, events,
            wf_all, wf_12, bls, oa, ba, pc, cm,
            p_val, tp, lp, ht, sb, brier, rpss_d, run_meta,
            pss_pers, pss_clim, fc_df, fc_ens, g1, g2, aucs)


# ===========================================================================
# 17.  DASHBOARD  (12 panels)
# ===========================================================================
def plot_dashboard(df, model, members, beliefs, b_std, disagr,
                   ewi, ewi_s, lo, hi, events,
                   wf_12, bls, oa, ba, pc, cm,
                   p_val, ht, sb, brier, rpss_d, run_meta,
                   pss_pers, pss_clim, fc_df, fc_ens, g1, g2, aucs,
                   out=None):
    if out is None:
        out = os.path.join(CFG["output_dir"], "figures",
                           "ocean_brain_v12_dashboard.png")

    dates = pd.to_datetime(df["Date"])
    STATES = PhysEMHMM.STATES
    C = {
        "bg":    "#04070d", "panel": "#091019",
        "grid":  "#141e2d", "border":"#1a2a3e",
        "text":  "#e8edf5", "muted": "#56677a",
        "ln":    "#22d3ee", "neu":   "#94a3b8",
        "el":    "#fb923c", "ewi":   "#a78bfa",
        "oni":   "#4ade80", "pdo":   "#f472b6",
        "warn":  "#fbbf24", "ok":    "#4ade80",
        "fail":  "#f87171", "ar1":   "#fb923c",
        "arima": "#38bdf8", "nmme":  "#34d399",
        "pers":  "#f472b6", "fc":    "#fde68a",
        "sub":   "#6ee7b7", "dis":   "#fbbf24",
    }
    cols = [C["ln"], C["neu"], C["el"]]

    def sa(ax, title=""):
        ax.set_facecolor(C["panel"])
        for sp in ax.spines.values():
            sp.set_color(C["border"]); sp.set_linewidth(0.5)
        ax.tick_params(colors=C["muted"], labelsize=7, length=2)
        ax.xaxis.label.set_color(C["muted"])
        ax.yaxis.label.set_color(C["muted"])
        ax.grid(axis="y", color=C["grid"], lw=0.3, alpha=0.5, zorder=0)
        ax.set_axisbelow(True)
        if title:
            ax.set_title(title, color=C["text"], fontsize=8.0,
                         fontweight="bold", pad=6, loc="left",
                         fontfamily="monospace")

    def evlines(ax):
        for i in events:
            ax.axvline(dates.iloc[i], color=C["warn"],
                       lw=0.45, alpha=0.28, zorder=1)

    def _ba(v):
        return float(np.mean([v[v["true"]==c]["correct"].mean()
                               for c in range(3)
                               if (v["true"]==c).sum() > 0]))
    def _pc(v, c):
        s = v[v["true"]==c]
        return float(s["correct"].mean()) if len(s) > 0 else 0.

    fig = plt.figure(figsize=(24, 24), facecolor=C["bg"])
    gs = gridspec.GridSpec(4, 3, figure=fig,
                           hspace=0.48, wspace=0.28,
                           left=0.048, right=0.978,
                           top=0.930, bottom=0.042)

    fig.text(0.50, 0.967,
             "ENSO-PDO  OCEAN BRAIN  v12.0  --  FINAL CANONICAL EDITION",
             ha="center", color=C["text"], fontsize=15,
             fontweight="bold", fontfamily="monospace")
    fig.text(0.50, 0.942,
             ("Ensemble-HMM | Kalman | Subsurface+Wind [L3] | "
              "Granger | ROC AUC | Event Strength | "
              "Balanced %.0f%% | p=%.4f | RPSS=%.3f" % (
                  ba*100, p_val, rpss_d["rpss"])),
             ha="center", color=C["muted"], fontsize=7.5)

    # P1: ONI + PDO + Subsurface proxy [N1]
    ax1 = fig.add_subplot(gs[0, :2])
    sa(ax1, "P1  ONI / PDO / Subsurface proxy [N1, L3]  1950-2024")
    oni_v = df["ONI_raw"].values; pdo_v = df["PDO_raw"].values
    sub_v = df["subsurface"].values
    ax1.axhline(0, color=C["grid"], lw=0.7)
    ax1.axhline(0.5, color=C["el"], lw=0.4, ls="--", alpha=0.4)
    ax1.axhline(-0.5, color=C["ln"], lw=0.4, ls="--", alpha=0.4)
    ax1.fill_between(dates, oni_v, 0, where=oni_v >= 0,
                     color=C["el"], alpha=0.12)
    ax1.fill_between(dates, oni_v, 0, where=oni_v < 0,
                     color=C["ln"], alpha=0.12)
    ax1.plot(dates, oni_v, color=C["oni"], lw=1.2, label="ONI (degC)")
    ax1.plot(dates, pdo_v, color=C["pdo"], lw=0.8, ls="--",
             alpha=0.65, label="PDO")
    ax1.plot(dates, sub_v*0.5, color=C["sub"], lw=0.7, ls=":",
             alpha=0.70, label="Subsurface proxy x0.5 (L3)")
    evlines(ax1)
    ax1.legend(loc="upper left", fontsize=7.0, facecolor=C["panel"],
               labelcolor=C["text"], framealpha=0.9, edgecolor=C["border"])

    # P2: ensemble beliefs + disagreement [N2]
    ax2 = fig.add_subplot(gs[1, :2])
    sa(ax2, "P2  Ensemble HMM beliefs (mean of 5 members) + disagreement [N2]")
    ax2.stackplot(dates, df["P_La"], df["P_Neutral"], df["P_El"],
                  colors=cols, alpha=0.80,
                  labels=["P(La Nina)",
                          "P(Neutral) ~%.0f%% OOS (L2)" % (pc["Neutral"]*100),
                          "P(El Nino)"])
    ax2b = ax2.twinx(); ax2b.set_facecolor("none")
    ax2b.tick_params(colors=C["muted"], labelsize=6)
    ax2b.spines["right"].set_color(C["border"])
    diss = gaussian_filter1d(disagr, sigma=6)
    ax2b.fill_between(dates, 0, diss, color=C["dis"],
                      alpha=0.25, label="Ensemble disagreement")
    ax2b.set_ylabel("Disagreement", fontsize=6.5, color=C["muted"])
    ax2b.set_ylim(0, 1)
    ax2.set_ylim(0, 1); evlines(ax2)
    ax2.legend(loc="upper left", fontsize=7.0, facecolor=C["panel"],
               labelcolor=C["text"], framealpha=0.9,
               edgecolor=C["border"], ncol=3)

    # P3: EWI
    ax3 = fig.add_subplot(gs[2, :2])
    sa(ax3, "P3  EWI (4-comp FIX2)  +  90% block-bootstrap CI (L8)")
    thr_v = float(ewi.mean() + ewi.std())
    ax3.fill_between(dates, lo, hi, color=C["ewi"], alpha=0.11, label="90% CI")
    ax3.plot(dates, ewi, color=C["ewi"], lw=0.6, alpha=0.18)
    ax3.plot(dates, ewi_s, color=C["ewi"], lw=1.9, label="EWI smoothed")
    ax3.axhline(thr_v, color=C["warn"], lw=0.9, ls="--", alpha=0.85,
                label="thr=%.2f" % thr_v)
    for i in events[:14]:
        ax3.axvline(dates.iloc[i], color=C["warn"], lw=0.45, alpha=0.28)
        ax3.annotate(str(int(df.iloc[i]["Year"])),
                     xy=(dates.iloc[i], ewi_s[i]),
                     xytext=(2, 4), textcoords="offset points",
                     color=C["warn"], fontsize=5.5, fontfamily="monospace")
    ax3.legend(loc="upper left", fontsize=7.5, facecolor=C["panel"],
               labelcolor=C["text"], framealpha=0.9, edgecolor=C["border"])

    # P4: Markov forecast + event strength [N3]
    ax4 = fig.add_subplot(gs[3, :2])
    sa(ax4, "P4  Markov forecast (L7) + Event Strength [N3] + Ensemble spread")
    fc_dates = pd.to_datetime(fc_df["Date"])
    ax4.fill_between(fc_dates, 0, fc_df["P_La"],
                     color=C["ln"], alpha=0.65, label="P(La Nina)")
    ax4.fill_between(fc_dates, fc_df["P_La"],
                     fc_df["P_La"]+fc_df["P_Neutral"],
                     color=C["neu"], alpha=0.65, label="P(Neutral)")
    ax4.fill_between(fc_dates,
                     fc_df["P_La"]+fc_df["P_Neutral"],
                     fc_df["P_La"]+fc_df["P_Neutral"]+fc_df["P_El"],
                     color=C["el"], alpha=0.65, label="P(El Nino)")
    ax4.plot(fc_dates, fc_df["conf"], color=C["fc"], lw=1.8,
             ls="--", label="Confidence")
    ax4b = ax4.twinx(); ax4b.set_facecolor("none")
    ax4b.tick_params(colors=C["muted"], labelsize=6)
    ax4b.spines["right"].set_color(C["border"])
    ax4b.fill_between(fc_dates, 0, fc_ens["std"],
                      color=C["sub"], alpha=0.30, label="Ens spread")
    ax4b.set_ylabel("Ens std", fontsize=6.5, color=C["muted"])
    ax4.set_ylim(0, 1.05)
    ax4.axhline(0.5, color=C["muted"], lw=0.7, ls=":", alpha=0.6)
    ax4.legend(loc="upper left", fontsize=7.0, facecolor=C["panel"],
               labelcolor=C["text"], framealpha=0.9,
               edgecolor=C["border"], ncol=2)
    for _, r in fc_df.iterrows():
        lbl = (r.strength
               .replace("Very Strong", "VS").replace("Strong", "St")
               .replace("Moderate", "Mo").replace("Weak", "Wk")
               .replace("El Nino", "EN").replace("La Nina", "LN")
               .replace("Neutral", "N"))
        ax4.text(r.Date, 0.02, lbl, ha="center", fontsize=5.2,
                 color=C["text"], fontfamily="monospace")

    # P5: Granger causality
    ax5 = fig.add_subplot(gs[0, 2])
    sa(ax5, "P5  Granger causality (L6)")
    lags = list(g1.keys())
    Fv1 = [g1[l][0] for l in lags]; pv1 = [g1[l][1] for l in lags]
    Fv2 = [g2[l][0] for l in lags]; pv2 = [g2[l][1] for l in lags]
    xp = np.arange(len(lags)); w = 0.35
    b1 = ax5.bar(xp-w/2, [min(F, 60) for F in Fv1], width=w,
                 color=C["pdo"], alpha=0.80, label="PDO->ONI")
    b2 = ax5.bar(xp+w/2, [min(F, 60) for F in Fv2], width=w,
                 color=C["oni"], alpha=0.80, label="ONI->PDO")
    for i, (p1, p2) in enumerate(zip(pv1, pv2)):
        if p1 < 0.001:
            ax5.text(i-w/2, b1[i].get_height()+0.5, "***",
                     ha="center", fontsize=6, color=C["text"])
        if p2 < 0.001:
            ax5.text(i+w/2, b2[i].get_height()+0.5, "***",
                     ha="center", fontsize=6, color=C["text"])
    ax5.set_xticks(xp)
    ax5.set_xticklabels(["lag%d" % l for l in lags],
                        fontsize=6.5, color=C["text"])
    ax5.set_ylabel("F-statistic", fontsize=7.5)
    ax5.legend(fontsize=6.5, facecolor=C["panel"],
               labelcolor=C["text"], framealpha=0.9, edgecolor=C["border"])
    ax5.text(0.02, 0.92, "*** p<0.001",
             transform=ax5.transAxes, fontsize=6.5,
             color=C["warn"], fontfamily="monospace")

    # P6: multi-horizon
    ax6 = fig.add_subplot(gs[1, 2])
    sa(ax6, "P6  Multi-horizon balanced accuracy")
    hs = ht["horizon"].values; ba_h = ht["balanced"].values
    ax6.plot(hs, ba_h, color=C["ewi"], lw=2.3, marker="o",
             ms=5, label="Balanced")
    ax6.plot(hs, ht["La Nina"].values, color=C["ln"],
             lw=1.1, ls="--", label="La Nina")
    ax6.plot(hs, ht["Neutral"].values, color=C["neu"],
             lw=1.1, ls=":", label="Neutral")
    ax6.plot(hs, ht["El Nino"].values, color=C["el"],
             lw=1.1, ls="--", label="El Nino")
    ax6.axhline(0.333, color=C["muted"], lw=0.8, ls=":", alpha=0.6)
    ax6.set_ylim(0, 1.08); ax6.set_xticks(hs)
    ax6.set_xlabel("Horizon (months)", fontsize=7.5)
    ax6.axvspan(3, 5, alpha=0.06, color=C["warn"])
    ax6.text(4, 0.04, "spring", color=C["warn"],
             fontsize=5.8, ha="center", fontfamily="monospace")
    ax6.legend(fontsize=6.5, facecolor=C["panel"],
               labelcolor=C["text"], framealpha=0.9,
               edgecolor=C["border"], ncol=2)

    # P7: model comparison bars
    ax7 = fig.add_subplot(gs[2, 2])
    sa(ax7, "P7  Balanced accuracy by class  (FIX3+FIX4+FIX6)")
    groups = ["Balanced", "La Nina", "Neutral", "El Nino"]
    m_list = [
        ("HMM-Ens v12", wf_12,         C["ewi"]),
        ("AR(1)",        bls["AR(1)"],  C["ar1"]),
        ("ARIMA",        bls["ARIMA"],  C["arima"]),
        ("NMME",         bls["NMME"],   C["nmme"]),
        ("Persist.",     bls["Persist"],C["pers"]),
    ]
    w = 0.14; x2 = np.arange(len(groups))
    for mi, (mname, mval, mcol) in enumerate(m_list):
        vals = [_ba(mval), _pc(mval,0), _pc(mval,1), _pc(mval,2)]
        ax7.bar(x2+(mi-2)*w, vals, width=w, color=mcol,
                alpha=0.82, label=mname)
    ax7.set_xticks(x2)
    ax7.set_xticklabels(groups, fontsize=7.5, color=C["text"])
    ax7.axhline(0.333, color=C["muted"], lw=0.8, ls=":", alpha=0.6)
    ax7.set_ylim(0, 1.18)
    ax7.legend(fontsize=6.0, facecolor=C["panel"],
               labelcolor=C["text"], framealpha=0.9,
               edgecolor=C["border"], ncol=2)
    auc_str = "  ".join("AUC-%s=%.3f" % (k[:2], v)
                        for k, v in aucs.items() if v == v)
    ax7.text(0.01, 0.88,
             "p=%.4f  PSS=%.3f  RPSS=%.3f\n%s" % (
                 p_val, pss_pers, rpss_d["rpss"], auc_str),
             transform=ax7.transAxes, fontsize=6.2, color=C["warn"],
             fontfamily="monospace",
             bbox=dict(boxstyle="round,pad=0.3", facecolor=C["panel"],
                       edgecolor=C["warn"], alpha=0.8))

    # P8: confusion matrix + status card
    ax8 = fig.add_subplot(gs[3, 2])
    ax8_cm = ax8.inset_axes([0.0, 0.44, 1.0, 0.56])
    ax8_cm.set_facecolor(C["panel"])
    for sp in ax8_cm.spines.values():
        sp.set_color(C["border"]); sp.set_linewidth(0.4)
    ax8_cm.tick_params(colors=C["muted"], labelsize=6.5)
    sa(ax8_cm, "P8  Confusion matrix  (OOS non-overlapping)")
    cm8 = LinearSegmentedColormap.from_list(
        "ht", ["#091019", "#fbbf24"], 256)
    ax8_cm.imshow(cm, cmap=cm8, aspect="auto")
    ax8_cm.set_xticks([0, 1, 2])
    ax8_cm.set_xticklabels(["LN", "N", "EN"], fontsize=7, color=C["text"])
    ax8_cm.set_yticks([0, 1, 2])
    ax8_cm.set_yticklabels(["LN", "N", "EN"], fontsize=7, color=C["text"])
    for i in range(3):
        for j in range(3):
            ax8_cm.text(j, i, str(cm[i, j]),
                        ha="center", va="center", fontsize=9,
                        fontweight="bold",
                        color="white" if cm[i, j] > cm.max()*0.45
                        else C["text"])
    ax8.axis("off")
    last = df.iloc[-1]
    di = int(np.argmax([last.P_La, last.P_Neutral, last.P_El]))
    ew = float(last.EWI)
    lv = "HIGH" if ew > 0.62 else "ELEVATED" if ew > 0.50 else "LOW"
    lvc = C["fail"] if ew > 0.62 else C["warn"] if ew > 0.50 else C["ok"]
    last_str = event_strength(float(last.ONI_raw))
    card = [
        ("STATUS  " + last.Date.strftime("%Y-%m"), C["text"], 9.5, True),
        ("ONI %+.2f  PDO %.2f" % (last.ONI_raw, last.PDO_raw),
         C["oni"], 8.0, False),
        ("Strength: " + last_str, C["warn"], 8.5, True),
        ("LN=%.2f  N=%.2f  EN=%.2f" % (
            last.P_La, last.P_Neutral, last.P_El),
         cols[di], 8.0, False),
        ("EWI %.3f  %s" % (ew, lv), lvc, 8.5, True),
        ("Regime: " + PhysEMHMM.STATES[di], cols[di], 9.0, True),
        ("BA=%.0f%%  p=%.4f" % (ba*100, p_val), C["warn"], 7.5, False),
        ("PSS=%.3f  RPSS=%.3f" % (pss_pers, rpss_d["rpss"]),
         C["ewi"], 7.5, False),
        ("Ens disagree=%.1f%%" % (disagr[-1]*100), C["dis"], 7.5, False),
    ]
    y = 0.40
    for txt, col, fs, bold in card:
        ax8.text(0.03, y, txt, color=col, fontsize=fs,
                 fontweight="bold" if bold else "normal",
                 transform=ax8.transAxes, fontfamily="monospace")
        y -= 0.044

    # P9: spring barrier
    ax9 = fig.add_subplot(gs[3, 0])
    sa(ax9, "P9  Spring predictability barrier")
    if len(sb) > 0:
        mo = sb["month"].values; sba = sb["ba"].values
        bc9 = [C["warn"] if m in [3, 4, 5] else C["ewi"] for m in mo]
        ax9.bar(mo, sba, color=bc9, alpha=0.75, width=0.7)
        ax9.axhline(0.333, color=C["muted"], lw=0.8, ls=":", alpha=0.6)
        ax9.set_xlim(0.5, 12.5); ax9.set_ylim(0, 1.05)
        ax9.set_xticks(range(1, 13))
        ax9.set_xticklabels(list("JFMAMJJASOND"),
                            fontsize=7, color=C["text"])
        ax9.text(4.0, 0.04, "spring", color=C["warn"],
                 fontsize=6, ha="center", fontfamily="monospace")

    # P10: reliability + ROC AUC [N4]
    ax10 = fig.add_subplot(gs[3, 1])
    sa(ax10, "P10  Reliability + ROC AUC [N4]  (FIX7 OOS)")
    ax10.plot([0, 1], [0, 1], color=C["muted"], lw=0.8, ls="--", alpha=0.5)
    for nm, col in zip(["La Nina", "Neutral", "El Nino"], cols):
        if nm not in brier: continue
        bd = brier[nm]
        if not bd["bin_centers"]: continue
        auc_v = aucs.get(nm, float("nan"))
        lbl = "%s  BS=%.3f" % (nm, bd["brier_raw"])
        if auc_v == auc_v:
            lbl += "  AUC=%.3f" % auc_v
        ax10.plot(bd["bin_centers"], bd["bin_fracs"],
                  color=col, lw=1.5, marker="o", ms=3, alpha=0.85,
                  label=lbl)
    ax10.set_xlim(0, 1); ax10.set_ylim(0, 1)
    ax10.set_xlabel("Forecast prob.", fontsize=7.5)
    ax10.set_ylabel("Observed freq.", fontsize=7.5)
    ax10.legend(fontsize=6.0, facecolor=C["panel"],
                labelcolor=C["text"], framealpha=0.9, edgecolor=C["border"])

    # P11: T matrix (inset under P5)
    ax11b = ax5.inset_axes([0.0, -0.62, 1.0, 0.52])
    ax11b.set_facecolor(C["panel"])
    ax11b.set_title("P11  Transition matrix T",
                    color=C["text"], fontsize=7.5,
                    fontweight="bold", pad=5, loc="left",
                    fontfamily="monospace")
    for sp in ax11b.spines.values():
        sp.set_color(C["border"]); sp.set_linewidth(0.4)
    ax11b.tick_params(colors=C["muted"], labelsize=6)
    cm11 = LinearSegmentedColormap.from_list(
        "oc", ["#091019", "#22d3ee"], 256)
    im11 = ax11b.imshow(model.T, cmap=cm11, vmin=0, vmax=1, aspect="auto")
    ax11b.set_xticks([0, 1, 2])
    ax11b.set_xticklabels(["LN", "N", "EN"], fontsize=7, color=C["text"])
    ax11b.set_yticks([0, 1, 2])
    ax11b.set_yticklabels(["LN", "N", "EN"], fontsize=7, color=C["text"])
    for i in range(3):
        for j in range(3):
            ax11b.text(j, i, "%.2f" % model.T[i, j],
                       ha="center", va="center", fontsize=8,
                       fontweight="bold",
                       color="white" if model.T[i, j] > 0.40
                       else C["text"])
    plt.colorbar(im11, ax=ax11b, fraction=0.046, pad=0.04)

    plt.savefig(out, dpi=150, bbox_inches="tight",
                facecolor=C["bg"], edgecolor="none")
    plt.close()
    print("  Dashboard ->", out)
    _show(out)
    return out


# ===========================================================================
# 18.  SAVE ALL ARTEFACTS  (directory structure from v3 pipeline) [N5]
# ===========================================================================
def save_all(df, wf_all, fc_df, brier, run_meta, ht,
             pss_pers, pss_clim, rpss_d, aucs):
    ts = time.strftime("%Y%m%d_%H%M%S")
    out = CFG["output_dir"]
    files = {}

    p = os.path.join(out, "data", "results_%s.csv" % ts)
    (df.assign(Date=df["Date"].dt.strftime("%Y-%m"))
       .drop(columns=["strength"], errors="ignore")
       .to_csv(p, index=False, float_format="%.4f"))
    files["results"] = p

    p = os.path.join(out, "data", "walkforward_%s.csv" % ts)
    (wf_all.assign(Date=pd.to_datetime(wf_all["Date"]).dt.strftime("%Y-%m"))
           .to_csv(p, index=False, float_format="%.4f"))
    files["walkforward"] = p

    p = os.path.join(out, "data", "forecast_%s.csv" % ts)
    (fc_df.assign(Date=pd.to_datetime(fc_df["Date"]).dt.strftime("%Y-%m"))
          .to_csv(p, index=False, float_format="%.4f"))
    files["forecast"] = p

    metrics = {
        "balanced_acc": run_meta["balanced_acc"],
        "p_value":      run_meta["p_value"],
        "pss_pers":     round(pss_pers, 4),
        "pss_clim":     round(pss_clim, 4),
        "rpss":         rpss_d.get("rpss"),
        "roc_auc":      {k: round(v, 4) for k, v in aucs.items()
                         if v == v},
        "horizon_table": ht.to_dict(orient="records"),
        "brier":         {k: {kk: round(vv, 4) if isinstance(vv, float)
                               else vv
                               for kk, vv in v.items()
                               if isinstance(vv, (float, int))}
                          for k, v in brier.items()},
    }
    p = os.path.join(out, "metrics", "metrics_%s.json" % ts)
    with open(p, "w") as f:
        json.dump(metrics, f, indent=2)
    files["metrics"] = p

    # SHA256 of results CSV [N5]
    csv_hash = sha256_csv(files["results"])
    files["sha256"] = csv_hash

    for k, v in files.items():
        if k != "sha256":
            print("  %-15s -> %s" % (k, v))
    return files


# ===========================================================================
# 19.  ENTRY POINT
# ===========================================================================
if __name__ == "__main__":
    import types

    # ------------------------------------------------------------------
    # CONFIGURATION
    # Set LIVE=True  to download fresh data from NOAA (needs internet)
    # Set FULL=True  for 200 bootstrap resamples (~8 min, publication)
    # Default: embedded data, 50 resamples (~2-3 min)
    # ------------------------------------------------------------------
    LIVE = False
    FULL = False
    # ------------------------------------------------------------------

    args = types.SimpleNamespace(live=LIVE, full=FULL)

    results = run_pipeline(live=args.live, fast=not args.full)

    (df, model, members, beliefs, b_std, disagr,
     ewi, ewi_s, lo, hi, events,
     wf_all, wf_12, bls, oa, ba, pc, cm,
     p_val, tp, lp, ht, sb, brier, rpss_d, run_meta,
     pss_pers, pss_clim, fc_df, fc_ens, g1, g2, aucs) = results

    files = save_all(df, wf_all, fc_df, brier, run_meta,
                     ht, pss_pers, pss_clim, rpss_d, aucs)

    plot_dashboard(df, model, members, beliefs, b_std, disagr,
                   ewi, ewi_s, lo, hi, events,
                   wf_12, bls, oa, ba, pc, cm,
                   p_val, ht, sb, brier, rpss_d, run_meta,
                   pss_pers, pss_clim, fc_df, fc_ens, g1, g2, aucs)

    print("\n" + "="*60)
    print("  HONEST POST  (38 words + hashtags)")
    print("="*60)
    print("""
  Physics-informed ENSO model built with AI. No domain training.
  Ensemble-HMM, Kalman state, event strength, 900-month NOAA data.
  Balanced %.0f%%. RPSS=%.3f. p=%.3f. Limitations stated. Code open.

  #ClimateScience #MachineLearning #HumanAI #ENSO
  #OpenScience #Python #HonestAI #DataScience
""" % (ba*100, rpss_d["rpss"], p_val))
    print("  Done.\n")
```
